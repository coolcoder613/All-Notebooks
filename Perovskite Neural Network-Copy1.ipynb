```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler,MinMaxScaler
from sklearn.compose import ColumnTransformer
```

```python
sb.violinplot(data = df, x = "heat_of_formation")
```

```python
df_num = df.drop("a_atom",axis =1)
df_num = df_num.drop("b1_atom",axis =1)
df_num = df_num.drop("b2_atom",axis =1)
df_num = df_num.drop("x_atom",axis =1)
scaler = MinMaxScaler(feature_range = (0,1))
num_pipeline = Pipeline([("scaler", scaler)])
```

```python
num_attribs = list(df_num)
num_cats = ["a_atom", "b1_atom", "b2_atom","x_atom"]

full_pipeline = ColumnTransformer([("cat", OneHotEncoder(), num_cats),("num", num_pipeline, num_attribs)])
```

```python
df_num
```

```python
df_final = full_pipeline.fit_transform(df)
```

```python
df_final
```

```python
df_final.shape
```

```python
X= np.delete(df_final, -1,1)
X= np.delete(X, -1,1)
Y= df_final[:,-2:]
```

```python
Y
```

```python
train_x = X[:400]
train_y = Y[:400]
```

```python
train_x.shape
```

```python
test_x = X[400:]
test_y = Y[400:]
```

```python
test_x.shape
```

```python
test_y.shape
```

```python
NN_model = Sequential()

NN_model.add(Dense(128, kernel_initializer='normal',input_dim = (39), activation='relu'))

NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))

NN_model.add(Dense(2, kernel_initializer='normal',activation='linear'))

NN_model.compile(loss='mse', optimizer='adam', metrics=['mse',"accuracy"])
NN_model.summary()
```

```python
history = NN_model.fit(train_x,train_y, validation_data = (test_x,test_y), epochs = 100)
```

```python
history.history["loss"]
history.history["val_loss"]
```
