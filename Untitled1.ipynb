```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("haberman.data")
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
for i in df:
    plt.figure(figsize=(3,3))
    plt.hist(df[i], bins = 10, color = 'pink', edgecolor='black', linewidth=1.0)
    plt.title(i)
    plt.show()
```

```python
plt.pie([list(df['Class']).count(1), list(df['Class']).count(2)], labels = ["Positive", "Negative"], autopct='%1.2f%%')
plt.show()
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth = 3)
```

```python
X = df.drop("Class",1)
y = df["Class"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
fn = ['Age', 'Year', 'Nodes']
cn = ["0", "1"]
fig, axis = plt.subplots(nrows = 1, ncols = 1, figsize=(3,3), dpi=300)
tree.plot_tree(clf, feature_names= fn, class_names=cn, filled = True)
```
